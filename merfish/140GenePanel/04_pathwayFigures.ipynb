```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from matplotlib import rc_context
```

# single cell type

```python
in_dir = r'Y:\coskun-lab\Zhou\12_MSG\MERSCOPE\20251229_msg-1_140gene\results\dann\common_genes\pathways\results'
fn_l = os.listdir(in_dir)
out_dir = r'Y:\coskun-lab\Zhou\12_MSG\MERSCOPE\20251229_msg-1_140gene\results\dann\common_genes\pathways\figures\single_cell_type'
```

```python
for fn in fn_l:
    data = pd.read_csv(os.path.join(in_dir, fn), sep='\t')
    top_data = data.sort_values("Combined Score", ascending=False).head(3)
    # top_data['Term'] = top_data['Term'] + ", Matching Score: " + top_data['Combined Score'].apply(lambda x: f"{x:.2f}")

    fig, ax = plt.subplots(1,1,figsize=(1.5,8.5))
    sns.barplot(
        data=top_data,
        y="Combined Score",
        x="Term",
        color="skyblue"
    )
    
    for container, (_, row) in zip(ax.containers[0], top_data.iterrows()):
        ax.text(
            container.get_x() + container.get_width()/2, # y-position at bar center
            0.02,                                       # x-position (a little inside the bar)
            row["Term"],                               # label text
            va="bottom", ha="center", fontsize=14, color="black", rotation=90
        )
    sns.despine(top=True, right=True)
    ax.set_yticklabels([])
    ax.set_ylabel("")
    ax.set_xticklabels([])
    ax.set_xlabel("")
    plt.tight_layout()
    plt.savefig(os.path.join(out_dir, f"{fn.split('.')[0]}.png"), dpi=300)
    plt.show()
```
