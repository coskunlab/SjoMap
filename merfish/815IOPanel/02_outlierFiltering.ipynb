```python
import numpy as np
import pandas as pd
import napari
import os
import tifffile
from skimage import io
import cv2
```

```python
results_dir = r'..\results'
```

```python
merscope_colors = pd.read_pickle(r'..\results\merscope_colors.pkl')
```

# R1 7263

```python
data = pd.read_csv(os.path.join(results_dir, r'202602051550_msg2-815IO_VMSC11512_region_R1-7263_annotated.csv'))
```

```python
data['center_y'] = - data['center_y']
```

```python
viewer = napari.Viewer()
for cell_type in data['cell_type'].unique():
    cell_data = data[data['cell_type'] == cell_type]
    points = cell_data[['center_x', 'center_y']].to_numpy()
    color = merscope_colors[cell_type]
    viewer.add_points(points, size=10, face_color=color, name=cell_type)
points = data[['center_x', 'center_y']].to_numpy()
viewer.add_points(points, size=5, face_color='white', name='all_cells')
```

```python
layer = viewer.layers['all_cells']
selected = layer.selected_data
```

```python
data_dropped = data.drop(index=selected)
```

```python
data_dropped.to_csv(os.path.join(results_dir, r'202602051550_msg2-815IO_VMSC11512_region_R1-7263_annotated_dfiltered.csv'), index=False)
```
