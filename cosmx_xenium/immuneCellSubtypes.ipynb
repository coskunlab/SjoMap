```python
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc_context
```

```python
adata = sc.read_h5ad(r'..\results\cell_type_annotation\v4\adata_83a.h5ad')
```

```python
sub_adata = adata[adata.obs['cell_types'].isin(['B-cell'])].copy()
```

```python
bcell_adata = sc.AnnData(sub_adata.X)
bcell_adata.obs_names = sub_adata.obs_names
bcell_adata.var_names = sub_adata.var_names
```

```python
sc.tl.umap(bcell_adata)
```

```python
sc.tl.leiden(bcell_adata, resolution=0.8)
```

```python
with rc_context({'figure.figsize': (5, 5)}):
    sc.pl.umap(bcell_adata, color='leiden', title='B-cell subtypes', save='_bcell_subtypes.png')
```

```python
sc.tl.rank_genes_groups(bcell_adata, 'leiden')
```

```python
with rc_context({'font.size': 12, 'figure.figsize': (4, 3)}):
    sc.pl.rank_genes_groups(bcell_adata, n_genes=25, groupby='leiden', ncols=3, save='_bcell_subtypes.png')
```

# combined analysis

```python
adata = sc.read_h5ad(r'..\results\cell_type_annotation\v4\adata_xenium_all_cells.h5ad')
```

```python
adata.obs['cell_types'].unique()
```

## b-cells

```python
sub_adata = adata[adata.obs['cell_types'].isin(['B-cell'])].copy()
```

```python
bcell_adata = sc.AnnData(sub_adata.X)
bcell_adata.obs_names = sub_adata.obs_names
bcell_adata.var_names = sub_adata.var_names
```

```python
sc.pp.neighbors(bcell_adata, n_neighbors=15)
```

```python
sc.tl.umap(bcell_adata)
```

```python
sc.tl.leiden(bcell_adata, resolution=0.5)
```

```python
with rc_context({'figure.figsize': (6, 6)}):
    sc.pl.umap(bcell_adata, color='leiden', legend_loc='on data', title='B-cell subtypes', save='_bcell_subtypes.png')
```

```python
sc.tl.rank_genes_groups(bcell_adata, 'leiden')
```

```python
with rc_context({'font.size': 12, 'figure.figsize': (4, 3)}):
    sc.pl.rank_genes_groups(bcell_adata, n_genes=20, groupby='leiden', ncols=3, save='_bcell_subtypes.png')
```

```python
# merge the subtype info back to the original adata
adata.obs["B-cell subtype"] = "other"
adata.obs.loc[bcell_adata.obs_names, "B-cell subtype"] = bcell_adata.obs["leiden"]
```

```python
sc.tl.rank_genes_groups(adata, 'B-cell subtype', key_added='rank_genes_bcell_subtype', references='other')
```

```python
with rc_context({'font.size': 12, 'figure.figsize': (4, 3)}):
    sc.pl.rank_genes_groups(adata, n_genes=20, key='rank_genes_bcell_subtype', groupby='B-cell subtype', ncols=3, save='_bcell_subtypes_all_cells.png')
```
