```python
import tifffile
import numpy as np
import pandas as pd
from skimage.filters import threshold_otsu
import matplotlib.pyplot as plt
from matplotlib import rc_context
from matplotlib.path import Path
import os
```

```python
out_dir = r'..\..\results\gat\stromal_pockets'
```

## 126-1

```python
count_df = pd.read_csv(r"Y:\coskun-lab2\Zhou\12_MSG\20250523_ro_lum\SSA-_126\counts_df.csv", index_col=0)
tissue_shape = pd.read_csv(r"Y:\coskun-lab2\Zhou\12_MSG\20250523_ro_lum\SSA-_126\tissue_shape.csv", index_col=0)
```

```python
# filter out count_df entries falls outside the tissue shape polygon
polygon_vertices = tissue_shape[['col', 'row']].to_numpy()
polygon_path = Path(polygon_vertices)
points = count_df[['col', 'row']].to_numpy()
inside_mask = polygon_path.contains_points(points)
count_df = count_df[inside_mask]
```

```python
thre = threshold_otsu(count_df['count'].to_numpy())
count_df['pocket'] = count_df['count'] > thre*1.5
```

```python
count_df['pocket'].value_counts()
```

```python
plt.scatter(count_df[count_df['pocket']==True]['row'], count_df[count_df['pocket']==True]['col'], color='r')
# plt.scatter(count_df[count_df['pocket']==False]['row'], count_df[count_df['pocket']==False]['col'], color='b')
```

```python
plt.scatter(count_df[count_df['pocket']==False]['row'], count_df[count_df['pocket']==False]['col'], color='b')
```

```python
stromal_pockets = count_df[count_df['pocket']==True]
stromal_pockets.to_csv(os.path.join(out_dir, '126-1_stromal.csv'))
```
